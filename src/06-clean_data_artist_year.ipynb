### Clean Data for further processing ###

```python
import pandas as pd

bremen_csv = "../output_csv/bremen_data_V4_translated.csv"

# Load and display the CSV as a table
df_bremen = pd.read_csv(bremen_csv)
print(f"The full dataset has {df_bremen.shape[0]} rows and {df_bremen.shape[1]} columns.")
print("First ten rows of the dataset")
df_bremen.head(10)
```

```python
df_bremen["birth year"]
```

```python
import re

# Clean birth year
df_bremen["birth year"] = df_bremen["birth year"].apply(
    lambda x: re.search(r"\d{4}", str(x)).group(0) if pd.notnull(x) and re.search(r"\d{4}", str(x)) else ""
)

df_bremen["birth year"]

```

```python
df_artists_summary = (
    df_bremen.groupby(['artistGNDId', 'artistName (preferred)', 'placeOfBirth', 'birth year'])
    .size() # Count the number of rows in each group
    .reset_index(name='num_works') # Convert the grouped result back to a DataFrame with a 'num_works' column
)

# Display the first few rows of the new summary DataFrame
print(df_artists_summary.head())

# Display the total number of unique artists
print(f"\nTotal unique artists: {len(df_artists_summary)}")
```

```python
# Export the cleaned and summarized data to a new CSV file
df_artists_summary.to_csv('../output_csv/bremen_data_v5_merge_artist.csv', index=False)

print("\nCleaned and merged data saved to bremen_data_v5_merge_artist.csv")
```
